1. Classes and Objects-
Class: A blueprint, template, or factory for creating something. It defines what data an entity will hold and what it can do, but it doesn't contain actual data itself.
Object: A real, physical instance built from that blueprint. If "Car" is the class, then the specific red sedan parked in your driveway is the object.

2. Constructors (`__init__`)-
The constructor is an automatic setup manager. The very instant you create an object from a class, this function wakes up behind the scenes to assign the initial values, configuration settings, or starting properties that the object needs to exist properly.

3. Instance Variables and Methods-
Instance Variable: Data that is completely unique to an individual object. For example, if you have two "User" objects, one might have the instance variable username set to "Alice" and the other to "Bob".
Instance Method: An action or behavior that an object can perform using its own unique data. It requires a reference to itself (written as `self` in Python) to look up or alter its own internal variables.

4. Class Variables-
A variable that is shared equally across **every single object** created from that class. It does not belong to a specific instance; it belongs to the blueprint itself. If you update a class variable, that change instantly reflects across all past and future objects of that class.

5. Encapsulation-
The protective capsule model. It is the practice of bundling data and the methods that modify that data into a single unit, while simultaneously locking down and shielding the internal components from direct outside tampering. External code must interact with the object through approved, controlled gateways rather than messing with its internal guts directly.

6. Inheritance-
The family tree concept. It allows a brand-new class (the child) to automatically inherit all the characteristics, data fields, and behaviors of an existing class (the parent). This eliminates copy-pasting code and allows you to build specialized versions of general concepts.

7. Polymorphism-
Meaning "many forms." It is the ability for completely different objects to respond to the exact same command or method name in their own unique way. For example, if you issue a "play" command to an Audio object and a Video object, both will start, but they execute the underlying action differently.

8. Abstraction-
The "need to know" rule. Abstraction hides complex, messy implementation details and only displays the essential, high-level interface. When you drive a car, you use the steering wheel and pedals (the abstraction); you do not need to know how the engine manages fuel injection ratios in real-time. In programming, it enforces strict structural design templates.

9. Method Overriding-
When a child class decides that the behavior it inherited from its parent class isn't quite right for its specific job. The child class rewrites that specific method with the exact same name, completely replacing or modifying the parent's original instructions when called on the child object.

10. `super()` Function-
A shortcut tool that allows a child class to reach back up the family tree and trigger a method or constructor belonging to its parent class. This ensures you can build on top of old logic instead of erasing it or rewriting it from scratch.

11. Access Modifiers-
Python does not enforce physical security barriers to block access to data, so developers rely on strict naming rules (using underscores) to signal how variables should be handled:

Public: Completely open. Anyone can view, read, or modify this variable from anywhere outside the class.
Protected (One Underscore): A gentle warning boundary. It tells other programmers, *"Please don't touch or change this variable unless you are working inside this class or one of its direct child classes."*
Private (Two Underscores): A hard locked restriction. It hides the variable so effectively that external code cannot easily see or access it directly, preventing accidental corruption of critical internal data.

```python
from abc import ABC, abstractmethod

# 1. ABSTRACTION (Subtask: Use abstraction)

class SecurityTool(ABC):
    def __init__(self, tool_name):
        self.tool_name = tool_name

    @abstractmethod
    def execute_scan(self, target):
        pass


# 2. INHERITANCE (Subtask: Implement inheritance)

class NetworkScanner(SecurityTool):
    
    # 3. CLASS VARIABLES
    scan_type = "Reconnaissance"

    # 4.(Subtask: Use constructors & Define attributes)
    def __init__(self, tool_name, speed_profile):
        super().__init__(tool_name)
        
        self.speed_profile = speed_profile
        
        # 5.(Subtask: Apply encapsulation)
        self.public_banner = f"--- {self.tool_name} Engine Active ---"
        self._protected_timeout = 30                    
        self.__discovered_vulnerabilities = []          

    # 6. METHODS (Subtask: Define methods)
    def get_vulnerabilities(self):
        return self.__discovered_vulnerabilities

    def _add_vulnerability(self, vuln_name):
        self.__discovered_vulnerabilities.append(vuln_name)

    def execute_scan(self, target):
        print(f"[{self.scan_type}] Initializing standard network sweep on {target} at speed T{self.speed_profile}...")
        self._add_vulnerability("Port 80 (HTTP) Open")



# 7. METHOD OVERRIDING (Subtask: Override methods)

class PortScanner(NetworkScanner):
    def __init__(self, tool_name, speed_profile, aggressive_mode=False):
        super().__init__(tool_name, speed_profile)
        self.aggressive_mode = aggressive_mode

    def execute_scan(self, target):
        print(f"[{self.scan_type}] Initializing Deep Port Scan on {target}...")
        if self.aggressive_mode:
            print("--> Running intensive OS fingerprinting and service detection (-A)...")
            self._add_vulnerability("SSH-2.0-OpenSSH_8.2p1 (Outdated)")
            self._add_vulnerability("CVE-2021-44228 (Log4j Risk)")
        else:
            print("--> Running stealth SYN stealth scan (-sS)...")
            self._add_vulnerability("Port 22 (SSH) Open")


# 8. POLYMORPHISM (Subtask: Demonstrate polymorphism)

class WebAppScanner(SecurityTool):
    def execute_scan(self, target):
        print(f"[Application Audit] Crawling web directories on {target}...")
        print("--> Injecting malicious payloads into input fields...")
        # Simulating finding a vulnerability
        return "SQL Injection detected on login route"


#(Subtask: Create classes and objects)

if __name__ == "__main__":
    print("=== ASSIGNMENT TASK EXECUTION ===\n")

    basic_recon = NetworkScanner(tool_name="NetSweep", speed_profile=3)
    stealth_nmap = PortScanner(tool_name="Nmap Engine", speed_profile=4, aggressive_mode=True)
    owasp_zap    = WebAppScanner(tool_name="OWASP ZAP")

    print(f"Initial Scan Type for NetSweep: {basic_recon.scan_type}")
    NetworkScanner.scan_type = "Active Penetration" # Modifying class variable
    print(f"Updated Scan Type for Nmap Engine: {stealth_nmap.scan_type}\n")

    audit_targets = [
        (basic_recon, "192.168.1.1"),
        (stealth_nmap, "10.0.0.55"),
        (owasp_zap, "https://test-target.com")
    ]

    for tool, target in audit_targets:
        print(tool.public_banner if hasattr(tool, 'public_banner') else f"--- {tool.tool_name} Active ---")
        tool.execute_scan(target)
        
        # Safe interaction with encapsulated data
        if hasattr(tool, 'get_vulnerabilities'):
            findings = tool.get_vulnerabilities()
            print(f"Audit Results from {tool.tool_name}: {findings}")
        print("-" * 50)
```

```python
#student management system
class Student:
    def __init__(self, name, roll, marks):
        self.name = name
        self.roll = roll
        self.marks = marks
    
    def calculate_grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        else:
            return "C"
    
    def display(self):
        print(f"Student: {self.name} | Roll No: {self.roll} | Grade: {self.calculate_grade()}")

s1 = Student("Soumalya", 101, 88)
s2 = Student("Sims", 102, 64)

s1.display()
s2.display()
```

```python
#bank acc system (encapsulation)
class BankAccount:
    def __init__(self, owner, initial_balance):
        self.owner = owner
        # Applying encapsulation via double underscore (Private Modifier)
        self.__balance = initial_balance 
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Successfully deposited ${amount}. Current Balance: ${self.__balance}")
        else:
            print("Deposit amount must be positive.")
    
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Successfully withdrew ${amount}. Remaining Balance: ${self.__balance}")
        elif amount > self.__balance:
            print(f"Transaction Denied: Insufficient funds for a ${amount} withdrawal.")
        else:
            print("Withdrawal amount must be positive.")
    
    def get_balance(self):
        return self.__balance

account = BankAccount("Soumalya", 500)
account.deposit(150)
account.withdraw(700)  # Should reject
account.withdraw(200)  # Should succeed
print(f"Final safe verification check balance: ${account.get_balance()}")
```

```python
#employe managment (inheritance)
class Employee:
    def __init__(self, name, emp_id):
        self.name = name
        self.emp_id = emp_id

    def work(self):
        print(f"Employee {self.name} (ID: {self.emp_id}) is working hard.")
 
class Manager(Employee):
    def __init__(self, name, emp_id, department):
        # Triggering parent constructor via super()
        super().__init__(name, emp_id)
        self.department = department

    def manage(self):
        print(f"Manager {self.name} is directing operations for the {self.department} team.")

mgr = Manager("Vikram", "MGR-404", "Cybersecurity Audit")
mgr.work()    
mgr.manage()  
```

```python
#vehicle system (polymorphism)
class Vehicle:
    def start(self):
        print("Vehicle starts moving.")
 
class Car(Vehicle):
    def start(self):
        print("Car engine purrs to life with an electronic key ignition switch.")
 
class Bike(Vehicle):
    def start(self):
        print("Bike engine roars to life using a manual kick or handle-bar self-start button.")

garage = [Car(), Bike(), Vehicle()]

for machine in garage:
    machine.start()
```

```python
#shape area calculator (abstraction)
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.l = length
        self.b = breadth
 
    def area(self):
        return self.l * self.b

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14159 * (self.radius ** 2)

rect = Rectangle(12, 5)
circ = Circle(7)

print(f"Rectangle Area Calculation: {rect.area()} sq units")
print(f"Circle Area Calculation: {circ.area():.2f} sq units")
```

```python
#library management(mini project)
class Book:
    def __init__(self, isbn, title, author):
        self.isbn = isbn
        self.title = title
        self.author = author
        self.is_issued = False

class Member:
    def __init__(self, member_id, name):
        self.member_id = member_id
        self.name = name
        self.borrowed_books = []  

class Library:
    def __init__(self):
        self.books_inventory = {}  
        self.members_registry = {} 

    def add_book(self, book):
        self.books_inventory[book.isbn] = book
        print(f"System: Book added successfully -> '{book.title}'")

    def register_member(self, member):
        self.members_registry[member.member_id] = member
        print(f"System: Member profile created -> {member.name}")

    def display_available_books(self):
        print("\n Current Available Books Inventory ")
        available = [b for b in self.books_inventory.values() if not b.is_issued]
        if not available:
            print("No books currently on shelves.")
        for b in available:
            print(f" ISBN: {b.isbn} | '{b.title}' by {b.author}")
        print("-" * 40)

    def issue_book(self, member_id, isbn):
        if member_id not in self.members_registry:
            print("Error: Missing member record.")
            return
        if isbn not in self.books_inventory:
            print("Error: Book not found in catalog.")
            return
        
        book = self.books_inventory[isbn]
        member = self.members_registry[member_id]

        if book.is_issued:
            print(f"Sorry, '{book.title}' is currently checked out by another member.")
        else:
            book.is_issued = True
            member.borrowed_books.append(book)
            print(f"Success: '{book.title}' has been checked out to {member.name}.")

    def return_book(self, member_id, isbn):
        if member_id not in self.members_registry:
            return
        
        member = self.members_registry[member_id]
        book_to_return = next((b for b in member.borrowed_books if b.isbn == isbn), None)

        if book_to_return:
            book_to_return.is_issued = False
            member.borrowed_books.remove(book_to_return)
            print(f"Success: '{book_to_return.title}' has been returned by {member.name}.")
        else:
            print("Error: This book configuration does not match the user's current lending record.")

# 1. Initialize central hub
city_library = Library()

# 2. Supply books
b1 = Book("978-1", "Black Hat Python", "Justin Seitz")
b2 = Book("978-2", "Count of Monte Cristo", "Alexandre Dumas")
city_library.add_book(b1)
city_library.add_book(b2)

# 3. Add users
m1 = Member("M-001", "Soumalya Jana")
city_library.register_member(m1)

# 4. Check inventory flow
city_library.display_available_books()
city_library.issue_book("M-001", "978-1")
city_library.display_available_books()

# 5. Try returning
city_library.return_book("M-001", "978-1")
city_library.display_available_books()
```
